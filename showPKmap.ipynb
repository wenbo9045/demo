```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline
%matplotlib
from mpl_toolkits.mplot3d import Axes3D

import json
```

```python
mapfile = "331/dataPkMap/Venue001.json"

with open(mapfile) as f:
  map = json.load(f)
# print(data)

%matplotlib
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

class Joint:
  
    def __init__(self, location, id='', laneids=[]):
        self.id = id
        self.location = location
        self.laneids = laneids
        
joints = {}
for joint in map['Joints']:
    laneids = []
    for laneid in joint['Link_IDs']:
        print(laneid['Link_ID'])
        laneids.append(laneid['Link_ID'])
    tempjoint = Joint(id=joint['ID'], location=joint['Location'], laneids=laneids)
#     ax.scatter(tempjoint.location['x'], tempjoint.location['y'], tempjoint.location['z'])
    joints[joint['ID']] = tempjoint


class Lane:
    
    def __init__(self, id, xs, ys, zs, fromjoint, gojoint):
        self.id = id
        self.xs = xs
        self.ys = ys
        self.zs = zs
        self.fromjoint = fromjoint
        self.gojoint = gojoint
        
lanes = {}
for lane in map['Links']:
    xs = []
    ys = []
    zs = []
    for waypoint in lane['Waypoints']:
        xs.append(waypoint['Location'][0])
        ys.append(waypoint['Location'][1])
        zs.append(waypoint['Location'][2])
    xs = np.array(xs)
    ys = np.array(ys)
    zs = np.array(zs)
#     print(lane['id'])
#     ax.plot(xs, ys, zs)
    gojoint = lane['Joint_IDs'][0]
    fromjoint = lane['Joint_IDs'][1]
    templane = Lane(lane['ID'], xs, ys, zs, fromjoint, gojoint)
    lanes[lane['ID']] = templane

joints_show = []
for joint in joints:
    if joint in joints_show:
        ax.scatter(joints[joint].location[0], joints[joint].location[1], joints[joint].location[2], marker='o', c='red')
    else:
        ax.scatter(joints[joint].location[0], joints[joint].location[1], joints[joint].location[2], marker='x', c='black')

lanes_show = []
for i in range(len(joints_show) - 1):
    jointid = joints_show[i]
#     print("from jointid: ", jointid)
    for laneid in joints[jointid].laneids:
        if lanes[laneid].gojoint == joints_show[i+1]:
            lanes_show.append(laneid)
lanes_show = []
for lane in lanes:
#     print(lane)
    
    if lane in lanes_show:
        ax.plot(lanes[lane].xs, lanes[lane].ys, lanes[lane].zs, c='red', linewidth=3)
    else:
        ax.plot(lanes[lane].xs, lanes[lane].ys, lanes[lane].zs, c='green')
    
    
mapfile = "331/dataPkMap/TrackMap_plan.json"
# TrackMap_plan.json为圆角图，现在用来全局规划
with open(mapfile) as f:
  map = json.load(f)
# print(data)


class Joint:
  
    def __init__(self, location, id='', laneids=[]):
        self.id = id
        self.location = location
        self.laneids = laneids
        
joints = {}
for joint in map['Joints']:
    laneids = []
    for laneid in joint['Link_IDs']:
        print(laneid['Link_ID'])
        laneids.append(laneid['Link_ID'])
    tempjoint = Joint(id=joint['ID'], location=joint['Location'], laneids=laneids)
#     ax.scatter(tempjoint.location['x'], tempjoint.location['y'], tempjoint.location['z'])
    joints[joint['ID']] = tempjoint


class Lane:
    
    def __init__(self, id, xs, ys, zs, fromjoint, gojoint):
        self.id = id
        self.xs = xs
        self.ys = ys
        self.zs = zs
        self.fromjoint = fromjoint
        self.gojoint = gojoint
        
lanes = {}
for lane in map['Links']:
    xs = []
    ys = []
    zs = []
    for waypoint in lane['Waypoints']:
        xs.append(waypoint['Location'][0])
        ys.append(waypoint['Location'][1])
        zs.append(waypoint['Location'][2])
    xs = np.array(xs)
    ys = np.array(ys)
    zs = np.array(zs)
#     print(lane['id'])
#     ax.plot(xs, ys, zs)
    gojoint = lane['Joint_IDs'][0]
    fromjoint = lane['Joint_IDs'][1]
    templane = Lane(lane['ID'], xs, ys, zs, fromjoint, gojoint)
    lanes[lane['ID']] = templane

joints_show = []
for joint in joints:
    if joint in joints_show:
        ax.scatter(joints[joint].location[0], joints[joint].location[1], joints[joint].location[2], marker='o', c='red')
    else:
        ax.scatter(joints[joint].location[0], joints[joint].location[1], joints[joint].location[2], marker='x', c='black')

lanes_show = []
for i in range(len(joints_show) - 1):
    jointid = joints_show[i]
#     print("from jointid: ", jointid)
    for laneid in joints[jointid].laneids:
        if lanes[laneid].gojoint == joints_show[i+1]:
            lanes_show.append(laneid)
lanes_show = []
for lane in lanes:
#     print(lane)
    
    if lane in lanes_show:
        ax.plot(lanes[lane].xs, lanes[lane].ys, lanes[lane].zs, c='red', linewidth=3)
    else:
        ax.plot(lanes[lane].xs, lanes[lane].ys, lanes[lane].zs, c='blue')
        
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
